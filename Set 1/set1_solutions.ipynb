```python
k = 9; v = [3, 6, 9, 6, 4, 2, 8]

def first_occurance_index(V):
    V_unique, V_indices = np.unique(V, return_index=True)
    return np.bincount(V_unique, weights=V_indices+1, minlength=k)[1:]

first_occurance_index(v)
```
